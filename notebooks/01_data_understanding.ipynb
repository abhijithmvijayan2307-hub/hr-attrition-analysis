# Phase 1 — Data Collection & Understanding

## Dataset
- Source: IBM HR Analytics (Kaggle)
- Shape: 1470 rows × 35 columns
- Target: Attrition (Yes/No)

## Key Findings
- Zero missing values
- 16.1% attrition rate → **imbalanced dataset** (will handle with SMOTE in Phase 3)
- Dropped 4 zero-variance columns: EmployeeCount, StandardHours, Over18, EmployeeNumber
- 24 numerical features, 7 categorical features

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\abhij\Desktop\hr-attrition-analysis\data\raw\WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Step 1 — Basic shape and info
print(df.shape)          # (1470, 35)
print(df.dtypes)
print(df.isnull().sum())  # Spoiler: no missing values!

# Step 2 — Drop useless columns (zero variance)
df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'],
        axis=1, inplace=True)

# Step 3 — Attrition rate
print(df['Attrition'].value_counts(normalize=True) * 100)
# Output: No: 83.9%   Yes: 16.1%  → Imbalanced dataset!

# Step 4 — Encode target
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Step 5 — Separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['str']).columns.tolist()

print("Numerical columns:", len(num_cols))
print("Categorical columns:", len(cat_cols))
print(cat_cols)
import os
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/hr_cleaned.csv', index=False)
print(" Cleaned data saved to data/processed/hr_cleaned.csv")
print(f"Final shape: {df.shape}")  # Should be (1470, 31) after dropping 4 columns
```
